# Exercise 3: Rotation

Rotate chain H by 90 around the axis defined by its first and last $\alpha$-carbon. Then apply this rotation with atom.set_coord(new_coord). Use nglview to visualize this rotation.

You can do this similar to the Glycin example. Define the axis, take care of the centering and apply the rotation.

Make sure to select the atoms of chain H via the Selection.unfold_entities() method.

```python
structure = parser.get_structure( "4XP1","4xp1.pdb")
model = structure[0]
chain_H = model["H"]

#get axis
atoms = Selection.unfold_entities(chain_H,"A")
dummy=0
for atom in atoms:
    if atom.get_id()=="CA":
        if dummy == 0:
            start = atom.get_vector()
            dummy = 1
        if dummy > 0:
            end =atom.get_vector()
#center at start
axis = end
#create rotmat
rot = rotaxis(np.pi/2,axis)
#apply rotations
for atom in atoms:
    at = atom.get_vector() - start
    at = at.left_multiply(rot)
    at = at + start
    atom.set_coord( np.array([at[i] for i in range(3)]))

import nglview as nv
view = nv.show_biopython(structure)
view.clear_representations()
view.add_ball_and_stick() #view as ball and stick 
view.clear_representations()
#add ribbons
view.add_cartoon('protein')
#add ball and stick for non-rotien
view.add_ball_and_stick('not protein')
view
```

# Exercise 6: Center of Mass

Write a function that calculates the center of Mass for a chain. 

Hint: atom._assign_atom_mass()

```python
def get_com(chain):
    atoms = chain.get_atoms()
    weighted_sum = np.zeros(3)
    total_mass= 0
    for atom in atoms:
        mass = atom._assign_atom_mass()
        total_mass += mass
        weighted_sum += mass * atom.get_coord()
    return weighted_sum / total_mass

#structure = parser.get_structure( "4XP1","4xp1.pdb")
model = structure[0]
com_A = get_com(model["A"])
print(com_A)
com_L = get_com(model["L"])
print(com_L)
com_H = get_com(model["H"])
print(com_H)
```
